```python
import pandas as pd
```

```python
input_csv = '/home/semyon/projects/vinbigdata/eda/dfs/result_faster_rcnn_24epoch.csv'
output_csv = input_csv + '__submission.csv'

train_csv = '/home/semyon/data/VinBigData/train.csv'
sample_csv = '/home/semyon/data/VinBigData/sample_submission.csv'
```

```python
tdf = pd.read_csv(train_csv)
idf = pd.read_csv(input_csv)
sdf = pd.read_csv(sample_csv)
odf = sdf.copy()
odf.set_index('image_id', drop=False, inplace=True)
# create class_id
# fix errors if any
class2id = tdf[['class_name', 'class_id']].groupby('class_name').mean().to_dict()['class_id']
id2class = {v:k for k,v in class2id.items()}
idf.class_id = idf.class_name.apply(lambda x: class2id[x] if x in class2id else 14)
idf.tail(2)
```

```python
for i, row in odf.iterrows():
    tmp = idf[idf.image_id == row.image_id]
    if len(tmp) == 0: # skip empty predictions
        continue
    prediction_str = ''
    for pre in tmp.itertuples():
        prediction_str += '{} {:.2f} {} {} {} {} '.format(pre.class_id, pre.rad_id, pre.x_min, pre.y_min, pre.x_max, pre.y_max)
#     print(prediction_str)
    odf.loc[row.image_id, 'PredictionString'] = prediction_str.strip()
```

```python
odf.reset_index(drop=True, inplace=True)
```

```python
odf.to_csv(output_csv, index=False)
```
