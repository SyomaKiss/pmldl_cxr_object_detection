```python
import subprocess
import tempfile
import pickle
```

```python
python = '/home/semyon/miniconda3/envs/open-mmlab/bin/python'
script = '/home/semyon/projects/mmdetection/tools/test.py'

CONFIG_FILE = '/home/semyon/projects/mmdetection/work_dirs/config_2x/config_2x.py'
CHECKPOINT_FILE = '/home/semyon/projects/mmdetection/work_dirs/config_2x/epoch_24.pth'
RESULT_FILE = tempfile.mktemp()+'.pkl'
EVAL_METRICS = 'bbox'
```

```python
cmd = '{python} {script} {CONFIG_FILE} {CHECKPOINT_FILE} --out {RESULT_FILE} --eval {EVAL_METRICS}'.format(
    python=python,
    script=script,
    CONFIG_FILE=CONFIG_FILE,
    CHECKPOINT_FILE=CHECKPOINT_FILE,
    RESULT_FILE=RESULT_FILE,
    EVAL_METRICS=EVAL_METRICS,)
```

```python
run = cmd.split()
out = subprocess.Popen(run, 
           stdout=subprocess.PIPE, 
           stderr=subprocess.STDOUT)

stdout,stderr = out.communicate()
# print(stdout.decode())
```
